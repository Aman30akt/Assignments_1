2. Explain the concept of function arguments and parameters in Python.
  - In Python, the terms "parameters" and "arguments" are distinct but related concepts concerning how data is passed into functions.
     Parameters:
      Parameters are variables defined within the parentheses in a function's definition. They act as placeholders for the values that the function expects to receive when it is called.
      Parameters specify the names and, optionally, the types of data that a function can accept. They define the "inputs" of the function.
     Arguments:
      Arguments are the actual values that are passed to a function when it is called. These values are assigned to the corresponding parameters within the function's scope.
      Arguments provide the concrete data that the function will operate on during its execution.

6. Explain the concept of generators in Python and how they are defined.
 - A generator in Python is a special type of function that allows you to iterate over a set of values. Unlike a regular function, a generator does not return its results all at once. Instead, it yields its values one by one, each time it is called. This makes it possible to generate an infinite sequence of values, as long as there is sufficient memory to store them.
 The syntax for creating a generator is similar to that of a regular function, with a few key differences. To create a generator, you use the yield keyword instead of the return keyword, and you wrap the function definition in parentheses instead of square brackets.
7. What are the advantages of using generators over regular functions?
 - Generators offer significant advantages over regular functions, primarily in memory efficiency and flexibility when dealing with large or potentially infinite sequences of data. Generators produce values on demand using the yield keyword, pausing execution and retaining state between calls, which contrasts with regular functions that execute once and return a single value or collection.

 Lambda functions are typically used when you need a short, simple function for a limited purpose, especially as an argument to other functions like map(), filter(), or sorted(). They help write concise code without the need to formally define a function with a name.
 However, for complex logic or longer code, using a regular function with def is clearer and better for readability.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - The map(), filter(), and reduce() functions in Python are built-in tools that facilitate functional programming paradigms, enabling concise and efficient data manipulation.
   1.  map():
    Purpose: Applies a given function to each item in an iterable (like a list, tuple, etc.) and returns a map object (an iterator in Python 3) containing the results.
    Transformation: It transforms each element individually, producing a new sequence of the same length as the input iterable, but with transformed values.
    Example: Squaring each number in a list
    2. filter():
    Purpose: Constructs an iterator from elements of an iterable for which a function returns True. It "filters" out elements that do not satisfy a given condition.
    Selection: It selects a subset of elements from the input iterable based on a boolean condition. The resulting sequence will typically be shorter than or equal to the length of the original.
    Example: Filtering even numbers from a list.
    3. reduce():
    Purpose: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It requires importing from the functools module.
    Aggregation: It aggregates all elements of an iterable into a single, cumulative result.
    Example: Calculating the sum of all elements in a list.

```python
 # 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
 def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
```

```python
#4.  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
for i in range(1, 201):
    if is_prime(i):
        print(i)
```

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b
fib = FibonacciIterator(10)

for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in power_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

with open('example.txt', 'w') as f:
    f.write("Hello, this is line 1.\n")
    f.write("This is line 2.\n")
    f.write("And this is line 3.\n")

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

file_name = 'example.txt'

for line in read_file_line_by_line(file_name):
    print(line.strip())
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 5), (3, 2), (2, 8), (5, 1)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temperatures = [25, 30, 15, 20]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
string = "Hello, World!"
vowels = 'aeiouAEIOU'
filtered_string = ''.join(filter(lambda x: x not in vowels, string))
print(filtered_string)
```

```python
# Data from the image
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

def process_orders(order_list):
    result = []
    for order in order_list:
        order_num = order[0]
        quantity = order[2]
        price_per_item = order[3]

        total = quantity * price_per_item
        if total < 100:
            total += 10
        result.append((order_num, round(total, 2)))
    return result

processed = process_orders(orders)
for item in processed:
    print(item)
```
